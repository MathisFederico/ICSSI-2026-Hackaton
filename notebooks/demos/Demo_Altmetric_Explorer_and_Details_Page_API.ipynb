# Altmetric Explorer API

This notebook provides a brief introduction and demonstration of how to use the Digital Science's Altmetric Explorer API.

```python
from getpass import getpass
import requests
import pandas as pd
import json

# Ask for API key (and API secret key, if needed)
api_key = getpass("Please enter your key for the Altmetric Explorer API: ")
# api_secret = getpass("Please enter your secret for the Author Check API: ")

# Pulling the exact query from Explorer API URL
api_url = f"https://www.altmetric.com/explorer/api/research_outputs?digest=04abf05c0434b88203910a57fe50c5a8c381de4d&filter%5Bq%5D=coral+reefs&key={api_key}"
```

There are 6 endpoints to the Explorer API:

*   research outputs
*   timeline
*   demographics
*   mentions
*   mention sources
*   journals

We will take a look at Research Outputs below.

## Research Outputs

```python
# function to pull data from the Explorer API using URLs
def get_api_data(api_url):

  response = requests.get(api_url)

  if response.status_code == 200:
    data = response.json()
    # Check if the response has a 'data' key and return its value, otherwise return the whole response
    if isinstance(data, dict) and 'data' in data:
      return data.get('data', [])
    else:
      return data # Return the data directly if not nested
  else:
    print(f"Error: Unable to retrieve data from the API. Status code: {response.status_code}")
    print(f"Response content: {response.text}")
    return None

# Example of how to use the function with the current api_url
research_outputs = get_api_data(api_url)

if research_outputs:
  print("Successfully retrieved data using the function:")
  display(research_outputs)
```

We can take the JSON data and put it into a more readable format.

```python
def extract_readable_data(research_outputs):

  data_for_df = []
  for item in research_outputs:
    attributes = item.get('attributes', {})
    data_for_df.append({
        'id': item.get('id'),
        'title': attributes.get('title'),
        'altmetric-score': attributes.get('altmetric-score'),
        'publication-date': attributes.get('publication-date'),
        'mentions': attributes.get('mentions')
    })
  return pd.DataFrame(data_for_df)


# Convert the list of research outputs (dictionaries) to a pandas DataFrame using the function
readable_df = extract_readable_data(research_outputs)

# Select and display the specified columns
display(readable_df)
```

## Mentions

Now we will take a look at the mentions endpoint. The URL below is from the same search query as above but pulling out the data associated under Mentions.



```python
api_url = f"https://www.altmetric.com/explorer/api/research_outputs/mentions?digest=04abf05c0434b88203910a57fe50c5a8c381de4d&filter%5Bq%5D=coral+reefs&key={api_key}"

mentions = get_api_data(api_url)
display(mentions[:10])
```

```python
def extract_mention_data(mentions_list):
    data_for_df = []
    for mention in mentions_list:
        attributes = mention.get('attributes', {})
        data_for_df.append({
            'title': attributes.get('title'),
            'posted-on': attributes.get('posted-on'),
            'post-type': attributes.get('post-type')
        })
    return pd.DataFrame(data_for_df)

# Convert the list of mentions (dictionaries) to a pandas DataFrame using the function
mentions_df = extract_mention_data(mentions)

# Display the DataFrame
display(mentions_df)
```

## Fetch Endpoint

```python
# fetch endpoint
api_url = f"https://api.altmetric.com/v1/fetch/id/4313515?key={details_api_key}"

mentions = get_api_data(api_url)
display(mentions)
```

```python
def extract_unique_users(mentions_data):
    user_data = []
    demographics = mentions_data.get('demographics', {})
    users = demographics.get('users', {})

    # Extract Twitter users
    twitter_users = users.get('twitter', {}).get('cohorts', {})
    for user_type, count in twitter_users.items():
        user_data.append({
            'platform': 'twitter',
            'user_type': user_type,
            'count': count
        })

    # Extract Mendeley users (by status and discipline) - unique_users are not directly available here
    # We can extract counts by status and discipline if needed, but not individual unique users
    # mendeley_users_status = users.get('mendeley', {}).get('by_status', {})
    # for status, count in mendeley_users_status.items():
    #     user_data.append({
    #         'platform': 'mendeley_status',
    #         'user_type': status,
    #         'count': count
    #     })

    # mendeley_users_discipline = users.get('mendeley', {}).get('by_discipline', {})
    # for discipline, count in mendeley_users_discipline.items():
    #     user_data.append({
    #         'platform': 'mendeley_discipline',
    #         'user_type': discipline,
    #         'count': count
    #     })

    # Extract other platforms from counts if unique_users are not directly available in 'users'
    counts = mentions_data.get('counts', {})
    platforms_with_unique_users_in_counts = ['news', 'blogs', 'facebook', 'googleplus', 'reddit', 'policy']

    for platform in platforms_with_unique_users_in_counts:
        platform_data = counts.get(platform, {})
        unique_users_list = platform_data.get('unique_users')
        if unique_users_list:
            for user in unique_users_list:
                 user_data.append({
                    'platform': platform,
                    'user_id': user,
                    'user_type': None # User type is not provided at this level
                })

    return pd.DataFrame(user_data)

# Assuming 'mentions' is the dictionary containing the Altmetric Details Page API response
if 'mentions' in locals() and isinstance(mentions, dict):
    unique_users_df = extract_unique_users(mentions)
    display(unique_users_df)
else:
    print("The 'mentions' variable is not available or is not in the expected format.")
```

```python
# function to extract tweeter_id's
def extract_tweeter_ids(mentions_data):
    tweeter_ids = []
    posts = mentions_data.get('posts', {})
    twitter_posts = posts.get('twitter', [])

    for post in twitter_posts:
        author = post.get('author', {})
        tweeter_id = author.get('tweeter_id')
        if tweeter_id:
            tweeter_ids.append(tweeter_id)

    return tweeter_ids

# Reminder: 'mentions' is the dictionary containing the Altmetric Details Page API response
if 'mentions' in locals() and isinstance(mentions, dict):
    all_tweeter_ids = extract_tweeter_ids(mentions)
    print("All Extracted Tweeter IDs:")
    display(all_tweeter_ids)
else:
    print("The 'mentions' variable is not available or is not in the expected format.")
```

We can also display the unique tweeter IDs and how many times each ID mentioned the research output of interest.

**Note**: Altmetric APIs do not provide information about tweets or tweeters, users will need to request that data directly from Twitter’s own API services. Users can utilize the Tweet IDs and User IDs that you obtain from Altmetric’s APIs to then query the Twitter API for this additional information. More information about this can be found on the [Altmetric API Documentation page](https://details-page-api-docs.altmetric.com/getting-started.html#limitations).

```python
# Convert the list of tweeter IDs into a pandas DataFrame
tweeter_ids_df = pd.DataFrame(all_tweeter_ids, columns=['tweeter_id'])

# Display the DataFrame
# display(tweeter_ids_df)

# Get unique tweeter IDs and their counts
tweeter_id_counts = tweeter_ids_df['tweeter_id'].value_counts()

# Display the counts
print("Unique Tweeter ID Counts:")
display(tweeter_id_counts)
```
